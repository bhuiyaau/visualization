```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50) #generates integer for 0-49
y = np.random.randint(0,100,50) #generate random int (y axis 0-100)
```

```python
fig, ax = plt.subplots(figsize=(5, 3)) #figure size, by inches
ax.scatter(x,y)
```

```python
fig,ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.hist(y) #counting the frequencies of the y (reason why x values not included)
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout() #displays cleaner
```

```python
font1={'family': 'sans-serif','color':'blue','size': 20}
font2={'family':'monospace','color':'green','size':14}
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time',fontdict=font1) #loc='right' <--- moves the position of the title
ax.set_ylabel('Total growth',fontdict=font2)
ax.set_xlabel('Years since start',fontdict=font2)
fig.tight_layout()
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.scatter(x,y,marker='*',color='indigo')
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.plot(x,y,marker='*',color='indigo',linestyle='--',linewidth=2)
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.grid(axis='y',color='blue',linewidth=2,linestyle='-.')
ax.plot(x,y,marker='*',markersize=12,color='#7425b9',linestyle='--',linewidth=2,markeredgecolor='#fa9359',markerfacecolor='#000000')
```
